## Correlation Analysis

```python
def corr_analysis(df):
  corr = df.corr(method='pearson')
  f, ax = plt.subplots(figsize=(6, 5))
  # Generate a mask for upper traingle
  mask = np.triu(np.ones_like(corr, dtype=bool))
  # Configure a custom diverging colormap
  cmap = sns.diverging_palette(230, 20, as_cmap=True)
  # Draw the heatmap
  sns.heatmap(corr, annot=True, mask = mask, cmap=cmap)
```

## Split and Reshape

```python
def split(df, train_size):
  df = df.astype(float)
  # split into train and test sets
  ts = int(len(df) * train_size)   # make 80% data train
  train = df[:ts] # set the train data
  test  = df[ts:] # set the test data 
 
  return train, test
```

```python
train, test = split(df_Ma, 0.8)
print("train shape", train.shape)
print("test shape", test.shape)
```

```python
train.head()
```

```python
def prepare(data):
  seq1 = data.Confirmed
  seq2 = data.Recovered
  seq3 = data.Deceased
  seq4 = data.New_Confirmerd
  seq5 = data.New_Recovered
  seq6 = data.New_Deaths
  out_seq = data.Confirmed

  seq1 = seq1.values.reshape((len(seq1), 1))
  seq2 = seq2.values.reshape((len(seq2), 1))
  seq3 = seq3.values.reshape((len(seq3), 1))
  seq4 = seq4.values.reshape((len(seq4), 1))
  seq5 = seq5.values.reshape((len(seq5), 1))
  seq6 = seq6.values.reshape((len(seq6), 1))
  out_seq = out_seq.values.reshape((len(out_seq), 1))
  dataset = np.hstack((seq1, seq2, seq3, seq4, seq5, seq6 , out_seq))
  return dataset
```

```python
prepare(train)
```

```python
# split a multivariate sequence into samples
def split_sequences(sequences, n_steps):
	X, y = list(), list()
	for i in range(len(sequences)):
		# find the end of this pattern
		end_ix = i + n_steps 
		# check if we are beyond the dataset
		if end_ix > len(sequences):
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequences[i:end_ix, :-1], sequences[end_ix-1, -1] 
		X.append(seq_x)
		y.append(seq_y)
	return np.array(X), np.array(y)
```

```python
# choose a number of time steps
n_steps = 5  # one week before
# convert into input/output
X_train, y_train = split_sequences(prepare(train), n_steps)
print(X_train.shape, y_train.shape)
# summarize the data
for i in range(len(X_train)):  # X shape [samples, time steps, features]
	print(X_train[i], y_train[i])
```
